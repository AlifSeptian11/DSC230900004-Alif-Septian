##Menyimpan data hasil cleansing pada SQLITE3

```python
#import library
import sqlite3
import pandas as pd
```

Connect to Database

```python
conn = sqlite3.connect("C:\\Users\\LENOVO\\Documents\\Binar_DSC\\Python\\API\\challange_gold.db")
print ("Opened database successfully")
```

##Create table input

```python
conn.execute('''
CREATE TABLE datacsv (
    tweet VARCHAR(13169) NOT NULL,
    HS INT,
    Abusive INT,
    HS_individual INT,
    HS_group INT,
    HS_religion INT,
    HS_race INT,
    HS_physical INT,
    HS_gender INT,
    HS_other INT,
    HS_weak INT,
    HS_moderate INT,
    HS_strong INT
);
''')
```

##Create table data output

```python
conn.execute('''
CREATE TABLE data_after_process (
    tweet VARCHAR(13169) NOT NULL,
    HS INT,
    Abusive INT,
    HS_individual INT,
    HS_group INT,
    HS_religion INT,
    HS_race INT,
    HS_physical INT,
    HS_gender INT,
    HS_other INT,
    HS_weak INT,
    HS_moderate INT,
    HS_strong INT
);
''')
```

```python
# read csv
df = pd.read_csv("data.csv", encoding="latin1")
df.head()
```

##Insert data into table input and output 

```python
# insert data to sqlite3 file input (data.csv)
table_name = "datacsv"
df.to_sql(table_name,conn,if_exists='append',index=False)
conn.commit()
```

```python
# show table file input (data.csv)
students = pd.read_sql_query('SELECT * FROM datacsv', conn)
students.head()
```

```python
# insert data to sqlite3 file output (data_afterpreprocessing.csv)
table_name = "data_after_process"
df.to_sql(table_name,conn,if_exists='append',index=False)
conn.commit()
```

```python
# show table file output 
students = pd.read_sql_query('SELECT * FROM data_after_process', conn)
students.head()
```
